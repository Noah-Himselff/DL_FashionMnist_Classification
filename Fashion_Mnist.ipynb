```python
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.model_selection import train_test_split
import seaborn as sns
%matplotlib inline
```

```python
fmtrain = pd.read_csv(r'C:\Users\user\Downloads\fashion-mnist_train.csv')
fmtest = pd.read_csv(r'C:\Users\user\Downloads\fashion-mnist_test.csv')
```

```python
fmtrain.head()
```

```python
fmtrain[:15]
```

```python

y=fmtrain['label']
```

```python
fmtrain.drop(['label'],axis=1,inplace=True)
x=fmtrain
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
y.info
```

```python
x.shape
```

```python
x.info
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(500,activation="selu"))
model.add(keras.layers.Dense(300,activation="selu"))
model.add(keras.layers.Dense(200,activation="selu"))
model.add(keras.layers.Dense(100,activation="selu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
model.layers
```

```python
earlystopping_callback = keras.callbacks.EarlyStopping(patience=5,
                             restore_best_weights=True)
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="adagrad",
              metrics=["accuracy"])
```

```python
callbacks=[earlystopping_callback]
```

```python
history = model.fit(x_train, y_train, epochs=50, validation_split=0.15)
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(history.history["loss"], label="train loss")
ax.plot(history.history["accuracy"], label="train accuracy")
ax.plot(history.history["val_loss"], label="validation loss")
ax.plot(history.history["val_accuracy"], label="validation accuracy")
ax.legend()
```

```python
ranking=history.history['accuracy']
ranking.sort(reverse=True)
```

```python
print('the best accuracy was',ranking[0])
```

```python
model.evaluate(x_test, y_test, verbose=0)
```
